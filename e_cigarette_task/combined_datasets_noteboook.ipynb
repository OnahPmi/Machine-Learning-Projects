```python
import pandas as pd
```

# Read the datasets into pandas dataframe

```python
df_dem =pd.read_csv("cleaned_data/cleaned_demography_data.csv")
df_can =pd.read_csv("cleaned_data/cleaned_cancer_data.csv")
df_cig =pd.read_csv("cleaned_data/cleaned_e-cigarettes_data.csv")
```

# Now, let's merge all the datasets on the `SEQN` column since it is the unique identifyer for the participants

```python
result1 = pd.merge(df_dem, df_can, how="inner", on="SEQN")
final_result = pd.merge(result1, df_cig, how="inner", on="SEQN")

final_result
```

## Recall that SMQ690H = E-cigarette use in the last 5 days

```python
e_cigarette_use = []

for num in final_result["SMQ690H"]:
    if num == 8:
        e_cigarette_use.append("Smoked")
    else:
        e_cigarette_use.append("Did not smoke")
```

```python
final_result["e_cigarette_use"] = e_cigarette_use
```

```python
final_result.drop("SMQ690H", axis=1, inplace=True)
```

```python
final_result["e_cigarette_use"].value_counts()
```
