<h2 style="text-align:center">Import necessary libraries</h2>

```python
import pandas as pd
import numpy as np
```

<h2 style="text-align:center">Dataset loading, and wrangling i.e., filling of missing values with each column's mean value</h2>

```python
column_names = ["FIPS", "State", "County", "Infant Motality Rate", "% Limited Access to Healthy Foods", "% Disconnected Youth",
               "Average Grade Performance in Reading Scores", "Average Grade Performance in Math Scores", 
                "% Enrolled in Free or Reduced Lunch", "Juvenile Arrest Rate", "% Broadband Acess"]

df = pd.read_excel("health factors.xlsx", header=[0,1],).iloc[:, :-1].copy()
df.columns = column_names

df1 = df.iloc[:,:3].copy() # A copy of the 1st to 3rd columns
df2 = df.iloc[:,3:].copy() # A copy of the 4st to last columns

df2 = df2.fillna(df2.mean()) #

new_df = pd.concat([df1, df2], axis=1).round(1)

# for i in ["Infant Motality Rate", "% Limited Access to Healthy Foods", 
#           "% Disconnected Youth", "% Enrolled in Free or Reduced Lunch", 
#           "Juvenile Arrest Rate", "% Broadband Acess"]:
#     new_df[i] = new_df[i].round().astype(int)

# for i in column_names:
#     if i == "FIPS" or i == "State" or i == "County":
#         continue
#     new_df[i] = new_df[i].round().astype(int)

for i in column_names[3:]:
    new_df[i] = new_df[i].round().astype(int)

# new_df
```

<h2 style="text-align:center">Sorting the Datasets</h2>

```python
sorted_df = new_df.sort_values(by=column_names[3 : ], ascending=True, inplace=False, kind='quicksort')
sorted_df["County Rank"] = [i for i in range(1, len(df)+1, 1)]
sorted_df.to_csv("sorted_health_factors.csv", index=False)
sorted_df
```
