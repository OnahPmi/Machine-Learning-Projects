# Import Pandas Library

```python
import pandas as pd
```

# Read the Extracted Features Datasets

```python
## Extracted Features for Cleaved octapeptide
amino_acid_binary_profile_cleaved = pd.read_csv("extracted_features_dataset/Amino acid Binary Profile (AABP)(cleaved).csv")
bond_composition_cleaved =pd.read_csv("extracted_features_dataset/Bond Composition (cleaved).csv")
physicochemical_properties_cleaved = pd.read_csv("extracted_features_dataset/Physico-Chemical Properties (cleaved).csv")

## Extracted Features for Uncleaved octapeptide
amino_acid_binary_profile_uncleaved = pd.read_csv("extracted_features_dataset/Amino acid Binary Profile (AABP)(uncleaved).csv")
bond_composition_uncleaved =pd.read_csv("extracted_features_dataset/Bond Composition (uncleaved).csv")
physicochemical_properties_uncleaved = pd.read_csv("extracted_features_dataset/Physico-Chemical Properties (uncleaved).csv")
```

# Combine the Extracted Features Datasets into one DF

```python
## Combining the datasets of cleaved octapeptides into one and relabelling it

cleaved_dataset = pd.concat([amino_acid_binary_profile_cleaved, bond_composition_cleaved, 
                             physicochemical_properties_cleaved.drop("PCP_SM", axis=1)], axis=1)
cleaved_dataset
```

```python
## Generating the labels for cleaved dataset

cleaved_label = pd.Series(["cleaved" for i in range(1001)], name="cleavage Status")
```

```python
cleaved_dataset_labelled = pd.concat([cleaved_dataset, cleaved_label], axis=1)
cleaved_dataset_labelled
```

```python
## Combining the datasets of non-cleaved octapeptides into one and relabelling it

uncleaved_dataset = pd.concat([amino_acid_binary_profile_uncleaved, bond_composition_uncleaved, 
                               physicochemical_properties_uncleaved], axis=1)
uncleaved_dataset
```

```python
## Generating the labels for non-cleaved dataset

uncleaved_label = pd.Series(["uncleaved" for i in range(4847)], name="cleavage Status")
```

```python
uncleaved_dataset_labelled = pd.concat([uncleaved_dataset, uncleaved_label], axis=1)
uncleaved_dataset_labelled
```

# Combine the cleaved_dataset_labelled and uncleaved_dataset_labelled

```python
combined_extracted_octapeptide_features = pd.concat([cleaved_dataset_labelled, uncleaved_dataset_labelled], axis=0)
combined_extracted_octapeptide_features
```

# Feature Selection of our combined_extracted_octapeptide_features

```python
## VarianceThreshold ised to remove low variance features from the feature variables
from sklearn.feature_selection import VarianceThreshold
```

```python
## Assigning the feature variables to X
X = combined_extracted_octapeptide_features.drop(["cleavage Status", "ID"], axis=1)
X
```

```python
selector = VarianceThreshold()
result = selector.fit_transform(X)
df =pd.DataFrame(result)
df
```

**Observe that there was no change in the dimension of the dataset indicating that no two features are the same**

# Drop the ID Column and Save the combined_extracted_octapeptide_features

```python
df_with_no_id = combined_extracted_octapeptide_features.drop("ID", axis=1)
df_with_no_id.to_csv("final_combined_extracted_features_dataset/final_combined_extracted_octapeptide_features.csv")
```
