## Create a mapping function for the "Your current year of Study" columns

```python
def map_study_year(data):
    study_year = []
    for item in data:
        if (item == "year 1" or item == "Year 1"):
            study_year.append(0)
        elif (item == "year 2" or item == "Year 2"):
            study_year.append(1)
        elif (item == "year 3" or item == "Year 3"):
            study_year.append(2)
        else:
            study_year.append(3)
    return study_year 
    
data = df["Your current year of Study"]
    
df["Your current year of Study"] = map_study_year(data)
df.head(2)
```

## Create a mapping function for the "Yes" - "No" and "Male" - "Female" questions

```python
def map_value(data):
        mylist = []
        for item in data:
            if item == "No" or item == "Female":
                mylist.append(0)
            else:
                mylist.append(1)
        return mylist

df["Do you have Depression?"] = map_value(df["Do you have Depression?"])
df
```
